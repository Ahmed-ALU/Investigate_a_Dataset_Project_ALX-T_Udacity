
> **The Second Step** into cleaning our data is to fix the issue of the multiple-variable cell, the ones that contains more than one value in the same cell, and we can find this in three columns ( 'director', 'genres', 'production_companies')
So in the next few cell we are going to check it and fix it

```python
# First we check the shape of the columns that contains hyprid data in :
hb_col= df[df['genres'].str.contains('[|]')]
hb_col
```

```python
#---Cleaning the hyprid data---
# in order to do that, we are going to create a python function to make it easy.

# used some help from the classroom
def fix_hyprid(base_column, split_columns, df) :

    hb = df[df[base_column].str.contains('[|]')]
    
    # create two copies of the hybrids dataframe
    df1 = hb.copy()
    df2 = hb.copy()

    # apply split function to each column of each dataframe copy
    for c in split_columns:
        df1[c] = df1[c].apply(lambda x: x.split("|")[0])
        try:
            df2[c] = df2[c].apply(lambda x: x.split("|")[1])
        except IndexError:
            pass
    
    # append the two dataframes
    new_rows = df1.append(df2)

    # drop each hybrid row from the original dataframe
    df.drop(hb.index, inplace=True)

    # append new_rows to dataframe
    df = df.append(new_rows, ignore_index=True)

    return df

```

**in order to fix the hyprid in all the columns we have (considering that we have multible hyprid in multible columns), we will have to repeat the function for each column everytime, and that will create copies everytime (identical but differnt in that column only), and we will base our analysis on it.**

```python
# call the function on the columns
df = fix_hyprid('genres', ['genres'], df)
df = fix_hyprid('production_companies', ['production_companies'], df)
df = fix_hyprid('director', ['director'], df);
```

### Research Question 2  (What kinds of properties are associated with movies that have high, low and moderatr Budget requirments?)

First of all, We will add an addition column with the budget_rate, as we will divide the budget into low, medium and high.

```python
# View the min, 25%, 50%, 75%, max pH values with Pandas describe
df['budget'].describe()
```

```python
# Bin edges that will be used to "cut" the data into groups
bin_edges = [ 1.000000e+00, 1.800000e+07, 3.465422e+07, 3.465423e+07, 4.250000e+08]
```

```python
# Labels for the four acidity level groups
bin_names = ['Low', 'Medium','Moderately', 'High']
```

```python
len(df.columns)
```

```python
# Creates budget_level column
df['budget_level'] = pd.cut(df['budget'], bin_edges, labels=bin_names)

# Checks for successful creation of this column
df.head()
```

```python
len(df.columns)
```

By now, we have added the extra column for the budget level so that we can act upon, now we will dicover the variables one by one deeper.

**First** We will check if there is a relation between the runtime and the budget level....

```python
# The mean runtime of each budget level
dfg = df.groupby('budget_level')['runtime'].mean()
dfg
```

```python
# Plotting the datat to visualize
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
levels = ['Low', 'Medium','Moderately', 'High']
students = [102.145962,110.087361,97.857539,114.240361]
ax.bar(levels,students)
plt.xlabel('budget level')
plt.ylabel('mean runtime')
plt.title('The mean runtime grouped by different budget levels')
plt.grid(True)
plt.show()
```

```python
plt.scatter(df['budget'], df['runtime'], label = "value01" )

plt.ylabel('Runtime')
plt.xlabel('Budget')

plt.title('Runtime vs Budget')
plt.legend()
plt.show()
```

**We see in the last figure that there is no big difference, but it also indicate that the higher the runtime the higher the possible budget is, in addition to that, it also indicates that the the lower the runtime is, the more moderate the possible budget will be. In between are the moderate rntime, whcih can have high and low possible budget.**
